A positive uplift value suggests the treatment is likely to increase the desired outcome. 
In our case, it means that a positive value indicates that the voter will vote democrat. 
The larger the number the more effective the treatment is assummed to be for that individual. 
A negative uplift value suggests the treatment might actually decrease the likelihood of the desired outcome and an uplift close to 0 implies that the treatment has little or no effect on the likelihood of the desired outcome.

```python
def classifyvoter(row):
    segments=[]
    if row['turnoutprediction']==1:
        segments.append('likelyvoter')
        
    if row['partisanshipprediction']==0:
        segments.append('persuadable')
        
    if row['turnoutprediction']==1 and row['partisanshipprediction'] in (1,2):
        segments.append('activesupporter')
    return ', '.join(segments)
```

```python
def assign_message(row):

    if pd.isna(row['partisanshipprediction']) or pd.isna(row['turnoutprediction']):
        return 'no message'
    if row['partisanshipprediction']==1 and row['turnoutprediction']==1:
        return 'message for likely partisanship and voter'
    elif row['partisanshipprediction']==1:
        return 'message for likely partisanship'
    elif row['turnoutprediction']==1:
        return 'message for likely voter'
    else:
        return 'general message'
```
